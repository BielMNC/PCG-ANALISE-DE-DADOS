```python
# IMPORTAMOS AS BIBLIOTECAS PANDAS E PLOTLY EXPRESS PARA REALIZAR A ANALISE DOS ARQUIVOS CSV E GERAR OS GRÁFICOS
import pandas as pd
import plotly.express as px
```

```python
# UTILIZAMOS ESTE COMANDO PARA FAZER A ABERTURA DE TODAS AS COLUNAS PARA FACILITAR A VISUALIZAÇÃO DOS DADOS
pd.set_option('display.max_columns', None)

# FIZEMOS A LEITURA DO ARQUIVO CSV DA CHAMPIONS LEAGUE COM O COMANDO "READ_CSV" DO PANDAS
champions = pd.read_csv('europe-uefa-champions-league-league-2022-to-2023-stats.csv')

# USAMOS O COMANDO "DROP" DO PANDAS PARA EXCLUIR AS COLUNAS QUE NÃO IRIAMOS UTILIZAR PARA NÃO NOS ATRAPALHAR
champions = champions.drop(['status', 'format', 'game_week', 'total_game_week', 'btts_percentage', 'over_05_percentage', 'over_15_percentage', 'over_25_percentage',
                            'over_35_percentage', 'over_45_percentage', 'over_55_percentage', 'under_05_percentage', 'under_15_percentage',
                            'under_25_percentage', 'under_35_percentage', 'under_45_percentage', 'under_55_percentage', 'over_65_corners_percentage', 'over_75_corners_percentage',
                            'over_85_corners_percentage', 'over_95_corners_percentage', 'over_105_corners_percentage', 'over_115_corners_percentage', 'over_125_corners_percentage',
                            'over_135_corners_percentage', 'goals_min_0_to_10', 'goals_min_11_to_20', 'goals_min_21_to_30', 'goals_min_31_to_40', 'goals_min_41_to_50', 'goals_min_51_to_60',
                            'goals_min_61_to_70', 'goals_min_71_to_80', 'goals_min_81_to_90', 'goals_min_0_to_15', 'goals_min_16_to_30', 'goals_min_31_to_45', 'goals_min_46_to_60',
                            'goals_min_61_to_75', 'goals_min_76_to_90'], axis=1)


# EXIBIMOS A TABELA DA CHAMPIONS LEAGUE
champions
```

```python
# FIZEMOS O MESMO PASSO A PASSO PARA O ARQUIVO CSV DA COPA LIBERTADORES
liberta = pd.read_csv('south-america-copa-libertadores-league-2023-to-2023-stats.csv')

liberta = liberta.drop(['status', 'format', 'game_week', 'total_game_week', 'btts_percentage', 'over_05_percentage', 'over_15_percentage', 'over_25_percentage',
                        'over_35_percentage', 'over_45_percentage', 'over_55_percentage', 'under_05_percentage', 'under_15_percentage',
                        'under_25_percentage', 'under_35_percentage', 'under_45_percentage', 'under_55_percentage', 'over_65_corners_percentage', 'over_75_corners_percentage',
                        'over_85_corners_percentage', 'over_95_corners_percentage', 'over_105_corners_percentage', 'over_115_corners_percentage', 'over_125_corners_percentage',
                        'over_135_corners_percentage', 'goals_min_0_to_10', 'goals_min_11_to_20', 'goals_min_21_to_30', 'goals_min_31_to_40', 'goals_min_41_to_50', 'goals_min_51_to_60',
                        'goals_min_61_to_70', 'goals_min_71_to_80', 'goals_min_81_to_90', 'goals_min_0_to_15', 'goals_min_16_to_30', 'goals_min_31_to_45', 'goals_min_46_to_60',
                        'goals_min_61_to_75', 'goals_min_76_to_90'], axis=1)

liberta
```

```python
# CRIAMOS UMA NOVA VARIÁVEL PARA ATRIBUIR A TABELA DA CHAMPIONS LEAGUE E DA LIBERTADORES JUNTAS
# PARA ISSO, USAMOS O COMANDO "CONCAT" DO PANDAS PARA JUNTAR AS DUAS TABELAS EM UMA SÓ
tabela_liga = pd.concat([champions, liberta], ignore_index=True)

tabela_liga # EXIBIMOS A TABELA DE AMBAS AS LIGAS JUNTAS
```

```python
# ATRIBUIMOS ESSES TEXTOS DENTRO DA VARIAVEL PARA UTILIZARMOS NOS GRÁFICOS
tabela_liga['Campeonato'] = ['Champions', 'Libertadores']
```

```python
# TOTAL DE JOGOS

# USAMOS O COMANDO "BAR" DO PLOTLY EXPRESS PARA CRIARMOS O GRAFICO DE BARRAS
fig = px.bar(tabela_liga, x=['Champions', 'Libertadores'], y='total_matches',
             color='Campeonato',
             labels={'x':'Campeonato', 'total_matches':''}, # COMANDO PARA RENOMEAR OS TEXTOS DO EIXO X E Y DO GRÁFICO
             color_discrete_sequence=[px.colors.qualitative.Dark24[22], px.colors.qualitative.Dark2[6]], # COMANDO PARA DEFINIR AS CORES DAS BARRAS
             text_auto=True) # COMANDO PARA EXIBIR OS DADOS DENTRO DAS SUAS RESPECTIVAS BARRAS DO GRÁFICO

fig.update_traces(textfont_size=15) # COMANDO PARA ALTERAR O TAMANHO DA FONTE DOS DADOS DE DENTRO DAS BARRAS DO GRÁFICO

# COMANDO "UPDATE_LAYOUT" DO PLOTLY EXPRESS PARA AJUSTAR OS DETALHES DE LAYOUT DO GRÁFICO
fig.update_layout(title={'text':'Total de Jogos', 'font': {'size': 25}}, # TITULO DO GRÁFICO
                  xaxis={'title': {'text': 'Campeonato', 'font': {'size': 20}}, 'tickfont': {'size': 17}}, # ALTERAR O TAMANHO DA FONTE DO EIXO X
                  legend={'title': {'font': {'size': 22}}, 'font': {'size': 19}}, # ALTERAR O TAMANHO DA FONTE DA LEGENDA
                  bargap=0.8) # DEFINIR A LARGURA DAS BARRAS DO GRÁFICO

fig.show() # EXIBIR O GRÁFICO
```

```python
# MEDIA DE CARTÕES POR JOGO

# USAMOS O COMANDO "BAR" DO PLOTLY EXPRESS PARA CRIARMOS O GRAFICO DE BARRAS
fig = px.bar(tabela_liga, x=['Champions', 'Libertadores'], y='average_cards_per_match',
             color='Campeonato',
             labels={'x':'Campeonato', 'average_cards_per_match':'Média'}, # COMANDO PARA RENOMEAR OS TEXTOS DO EIXO X E Y DO GRÁFICO
             color_discrete_sequence=[px.colors.qualitative.Dark24[22], px.colors.qualitative.Dark2[6]], # COMANDO PARA DEFINIR AS CORES DAS BARRAS
             text_auto=True) # COMANDO PARA EXIBIR OS DADOS DENTRO DAS SUAS RESPECTIVAS BARRAS DO GRÁFICO

fig.update_traces(textfont_size=15) # COMANDO PARA ALTERAR O TAMANHO DA FONTE DOS DADOS DE DENTRO DAS BARRAS DO GRÁFICO

# COMANDO "UPDATE_LAYOUT" DO PLOTLY EXPRESS PARA AJUSTAR OS DETALHES DE LAYOUT DO GRÁFICO
fig.update_layout(title={'text':'Média de Cartões por Jogo', 'font': {'size': 25}}, # TITULO DO GRÁFICO
                  xaxis={'title': {'text': 'Campeonato', 'font': {'size': 20}}, 'tickfont': {'size': 17}}, # ALTERAR O TAMANHO DA FONTE DO EIXO X
                  yaxis={'title': {'text': 'Média', 'font': {'size': 20}}}, # ALTERAR O TAMANHO DA FONTE DO EIXO Y
                  legend={'title': {'font': {'size': 22}}, 'font': {'size': 19}}, # ALTERAR O TAMANHO DA FONTE DA LEGENDA
                  bargap=0.8) # DEFINIR A LARGURA DAS BARRAS DO GRÁFICO

fig.show() # EXIBIR O GRÁFICO
```

```python
# CLEAN SHEETS

# USAMOS O COMANDO "MELT" DO PANDAS PARA MODIFICAR A TABELA DO FORMATO LARGO PARA O FORMATO LONGO E DEIXANDO OS IDENTIFICADORES DEFINIDOS.
tabela_long = tabela_liga.melt(id_vars='Campeonato', value_vars=['total_matches', 'clean_sheets_percentage'], var_name='Variável', value_name='Quantidade')

# COMANDO "REPLACE" DO PANDAS PARA RENOMEAR OS TEXTOS
tabela_long['Variável'] = tabela_long['Variável'].replace({
    'total_matches':'Total de jogos',
    'clean_sheets_percentage':'Porcentagem de jogos que um time não sofreu gol'
})

# USAMOS O COMANDO "BAR" DO PLOTLY EXPRESS PARA CRIARMOS O GRAFICO DE BARRAS
fig = px.bar(tabela_long, x='Campeonato', y='Quantidade', color='Variável', barmode='group',
             color_discrete_sequence=[px.colors.qualitative.Dark24[22], px.colors.qualitative.Dark2[6]], # COMANDO PARA DEFINIR AS CORES DAS BARRAS
             text_auto=True) # COMANDO PARA EXIBIR OS DADOS DENTRO DAS SUAS RESPECTIVAS BARRAS DO GRÁFICO

fig.update_traces(textfont_size=15) # COMANDO PARA ALTERAR O TAMANHO DA FONTE DOS DADOS DE DENTRO DAS BARRAS DO GRÁFICO

fig.update_layout(title={'text':'Porcentagem de jogos que um time não sofreu gol', 'font': {'size': 25}}, # TITULO DO GRÁFICO
                  xaxis={'title': {'text': 'Campeonato', 'font': {'size': 20}}, 'tickfont': {'size': 17}}, # ALTERAR O TAMANHO DA FONTE DO EIXO X
                  yaxis={'title': {'text': 'Quantidade', 'font': {'size': 20}}}, # ALTERAR O TAMANHO DA FONTE DO EIXO Y
                  legend={'title': {'font': {'size': 22}}, 'font': {'size': 19}}, # ALTERAR O TAMANHO DA FONTE DA LEGENDA
                  bargap=0.6) # DEFINIR A LARGURA DAS BARRAS DO GRÁFICO

fig.show() # EXIBIR O GRÁFICO
```

```python
pd.set_option('display.max_rows', None) # UTILIZAMOS ESTE COMANDO PARA FAZER A ABERTURA DE TODAS AS LINHAS PARA FACILITAR A VISUALIZAÇÃO DOS DADOS

# FIZEMOS A LEITURA DO ARQUIVO CSV DE TIMES DA CHAMPIONS LEAGUE COM O COMANDO "READ_CSV" DO PANDAS
time_champions = pd.read_csv('europe-uefa-champions-league-teams-2022-to-2023-stats.csv')

# EXIBIMOS A TABELA DE TIMES DA CHAMPIONS LEAGUE
time_champions
```

```python
index_to_keep = 66 # USAMOS ESTE COMANDO PARA MANTER O INDEX QUE IREMOS REALIZAR A ANALISE DOS DADOS

time_champions = time_champions.loc[[index_to_keep]]

time_champions # EXIBIMOS A TABELA DE TIMES DA CHAMPIONS LEAGUE COM APENAS O INDEX DEFINIDO
```

```python
# FIZEMOS A LEITURA DO ARQUIVO CSV DE TIMES DA LIBERTADORES COM O COMANDO "READ_CSV" DO PANDAS
time_liberta = pd.read_csv('south-america-copa-libertadores-teams-2023-to-2023-stats.csv')

# EXIBIMOS A TABELA DE TIMES DA LIBERTADORES
time_liberta
```

```python
index_to_keep = 20 # USAMOS ESTE COMANDO PARA MANTER O INDEX QUE IREMOS REALIZAR A ANALISE DOS DADOS

time_liberta = time_liberta.loc[[index_to_keep]]

time_liberta # EXIBIMOS A TABELA DE TIMES DA LIBERTADORES COM APENAS O INDEX DEFINIDO
```

```python
# ATRIBUÍMOS EM UMA NOVA VARIÁVEL AS DUAS TABELAS DE TIMES JUNTAS QUE USAMOS O COMANDO "CONCAT" DO PANDAS PARA JUNTAR
tabela_times = pd.concat([time_champions, time_liberta], ignore_index=True)

tabela_times # EXIBIMOS A TABELA DE TIMES AMBAS AS LIGAS JUNTAS
```

```python
# ATRIBUIMOS ESSES TEXTOS DENTRO DA VARIAVEL PARA UTILIZARMOS NOS GRÁFICOS
tabela_times['Times'] = ['Manchester City', 'Fluminense']
```

```python
# VITORIAS

# USAMOS O COMANDO "MELT" DO PANDAS PARA MODIFICAR A TABELA DO FORMATO LARGO PARA O FORMATO LONGO E DEIXANDO OS IDENTIFICADORES DEFINIDOS.
tabela_long = tabela_times.melt(id_vars='Times', value_vars=['wins', 'wins_home', 'wins_away'], var_name='Variável', value_name='Quantidade')

# COMANDO "REPLACE" DO PANDAS PARA RENOMEAR OS TEXTOS
tabela_long['Variável'] = tabela_long['Variável'].replace({
    'wins':'Vitórias',
    'wins_home':'Vitórias em casa',
    'wins_away':'Vitórias fora de casa'
})

# USAMOS O COMANDO "BAR" DO PLOTLY EXPRESS PARA CRIARMOS O GRAFICO DE BARRAS
fig = px.bar(tabela_long, x='Times', y='Quantidade', color='Variável', barmode='group',
             color_discrete_sequence=[px.colors.qualitative.Dark24[5], px.colors.qualitative.Alphabet[4], px.colors.qualitative.Set1[8]], # COMANDO PARA DEFINIR AS CORES DAS BARRAS
             text_auto=True) # COMANDO PARA EXIBIR OS DADOS DENTRO DAS SUAS RESPECTIVAS BARRAS DO GRÁFICO

fig.update_traces(textfont_size=15) # COMANDO PARA ALTERAR O TAMANHO DA FONTE DOS DADOS DE DENTRO DAS BARRAS DO GRÁFICO

fig.update_layout(title={'text':'Análise de Vitórias', 'font': {'size': 25}}, # TITULO DO GRÁFICO
                  xaxis={'title': {'text': 'Times', 'font': {'size': 20}}, 'tickfont': {'size': 17}}, # ALTERAR O TAMANHO DA FONTE DO EIXO X
                  yaxis={'title': {'text': 'Quantidade', 'font': {'size': 20}}}, # ALTERAR O TAMANHO DA FONTE DO EIXO Y
                  legend={'title': {'font': {'size': 22}}, 'font': {'size': 19}}, # ALTERAR O TAMANHO DA FONTE DA LEGENDA
                  bargap=0.5) # DEFINIR A LARGURA DAS BARRAS DO GRÁFICO

fig.show() # EXIBIR O GRÁFICO
```

```python
# GOLS

# USAMOS O COMANDO "MELT" DO PANDAS PARA MODIFICAR A TABELA DO FORMATO LARGO PARA O FORMATO LONGO E DEIXANDO OS IDENTIFICADORES DEFINIDOS.
tabela_long = tabela_times.melt(id_vars='Times', value_vars=['goals_scored', 'goals_conceded'], var_name='Variável', value_name='Quantidade')

# COMANDO "REPLACE" DO PANDAS PARA RENOMEAR OS TEXTOS
tabela_long['Variável'] = tabela_long['Variável'].replace({
    'goals_scored':'Gols marcados',
    'goals_conceded':'Gols sofridos'})

# USAMOS O COMANDO "BAR" DO PLOTLY EXPRESS PARA CRIARMOS O GRAFICO DE BARRAS
fig = px.bar(tabela_long, x='Times', y='Quantidade', color='Variável', barmode='group',
             color_discrete_sequence=[px.colors.qualitative.Dark24[5], px.colors.qualitative.Alphabet[4]], # COMANDO PARA DEFINIR AS CORES DAS BARRAS
             text_auto=True) # COMANDO PARA EXIBIR OS DADOS DENTRO DAS SUAS RESPECTIVAS BARRAS DO GRÁFICO

fig.update_traces(textfont_size=15) # COMANDO PARA ALTERAR O TAMANHO DA FONTE DOS DADOS DE DENTRO DAS BARRAS DO GRÁFICO

fig.update_layout(title={'text':'Análise de Gols', 'font': {'size': 25}}, # TITULO DO GRÁFICO
                  xaxis={'title': {'text': 'Times', 'font': {'size': 20}}, 'tickfont': {'size': 17}}, # ALTERAR O TAMANHO DA FONTE DO EIXO X
                  yaxis={'title': {'text': 'Quantidade', 'font': {'size': 20}}}, # ALTERAR O TAMANHO DA FONTE DO EIXO Y
                  legend={'title': {'font': {'size': 22}}, 'font': {'size': 19}}, # ALTERAR O TAMANHO DA FONTE DA LEGENDA
                  bargap=0.6) # DEFINIR A LARGURA DAS BARRAS DO GRÁFICO

fig.show() # EXIBIR O GRÁFICO
```

```python
# CARTÕES E FALTAS

# USAMOS O COMANDO "MELT" DO PANDAS PARA MODIFICAR A TABELA DO FORMATO LARGO PARA O FORMATO LONGO E DEIXANDO OS IDENTIFICADORES DEFINIDOS.
tabela_long = tabela_times.melt(id_vars='Times', value_vars=['cards_total', 'fouls'], var_name='Variável', value_name='Quantidade')

# COMANDO "REPLACE" DO PANDAS PARA RENOMEAR OS TEXTOS
tabela_long['Variável'] = tabela_long['Variável'].replace({
    'cards_total':'Total de cartões',
    'fouls':'Faltas'})

# USAMOS O COMANDO "BAR" DO PLOTLY EXPRESS PARA CRIARMOS O GRAFICO DE BARRAS
fig = px.bar(tabela_long, x='Times', y='Quantidade', color='Variável', barmode='group',
             color_discrete_sequence=[px.colors.qualitative.Dark24[5], px.colors.qualitative.Alphabet[4]], # COMANDO PARA DEFINIR AS CORES DAS BARRAS
             text_auto=True) # COMANDO PARA EXIBIR OS DADOS DENTRO DAS SUAS RESPECTIVAS BARRAS DO GRÁFICO

fig.update_traces(textfont_size=15) # COMANDO PARA ALTERAR O TAMANHO DA FONTE DOS DADOS DE DENTRO DAS BARRAS DO GRÁFICO

fig.update_layout(title={'text':'Faltas e Cartões', 'font': {'size': 25}}, # TITULO DO GRÁFICO
                  xaxis={'title': {'text': 'Times', 'font': {'size': 20}}, 'tickfont': {'size': 17}}, # ALTERAR O TAMANHO DA FONTE DO EIXO X
                  yaxis={'title': {'text': 'Quantidade', 'font': {'size': 20}}}, # ALTERAR O TAMANHO DA FONTE DO EIXO Y
                  legend={'title': {'font': {'size': 22}}, 'font': {'size': 19}}, # ALTERAR O TAMANHO DA FONTE DA LEGENDA
                  bargap=0.6) # DEFINIR A LARGURA DAS BARRAS DO GRÁFICO

fig.show() # EXIBIR O GRÁFICO
```
